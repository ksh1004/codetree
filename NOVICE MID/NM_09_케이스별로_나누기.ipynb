두 선분
- https://inha.codetree.ai/ko/trails/complete/curated-cards/intro-two-lines/description

```python
x1, x2, x3, x4 = map(int, input().split())

if(x2 < x3 or x4 < x1):
    print('nonintersecting')
else:
    print('intersecting')
```

두 직사각형
- https://inha.codetree.ai/ko/trails/complete/curated-cards/intro-two-rectangles/description

```python
x1, y1, x2, y2 = map(int, input().split())
a1, b1, a2, b2 = map(int, input().split())

if(x2 < a1 or a2 < x1 or b2 < y1 or y2 < b1):
    print('nonoverlapping')
else:
    print('overlapping')
```

구역 청소
- https://inha.codetree.ai/ko/trails/complete/curated-cards/challenge-district-cleaning/description

```python
a, b = map(int, input().split())
c, d = map(int, input().split())
sep = [0 for _ in range(101)]

for i in range(a, b):
    sep[i] = 1
for j in range(c, d):
    sep[j] = 1

cnt = 0
for k in range(0, 101):
    if(sep[k] == 1):
        cnt += 1

print(cnt)
```

전부 겹치는 선분
- https://inha.codetree.ai/ko/trails/complete/curated-cards/challenge-overlapping-line-segments/description

```python
import sys

INT_MAX = sys.maxsize

n = int(input())

max_x1 = 0
min_x2 = INT_MAX

# 시작점 중 가장 뒤에 있는 좌표와 끝점 중 가장 앞에 있는 점의 좌표를 확인합니다.
for _ in range(n):
    x1, x2 = tuple(map(int, input().split()))
    max_x1 = max(max_x1, x1)
    min_x2 = min(min_x2, x2)

# 만약 어느 한 선분이라도 시작점이 다른 선분의 끝점보다 뒤에 온다면
# 선분이 전부 겹칠 수 없습니다.
if min_x2 >= max_x1:
    print("Yes")
else:
    print("No")
```

전부 겹치는 선분 2
- https://inha.codetree.ai/ko/trails/complete/curated-cards/test-overlapping-line-segments-2/description

```python
N = int(input())
grid = [list(map(int, input().split())) for _ in range(N)]

answer = False

for i in range(N):
    sep = [0 for _ in range(101)]
    for j in range(N):
        if(i == j):
            continue
        else:
            for k in range(grid[j][0], grid[j][1] + 1):
                sep[k] += 1
    for j in range(101):
        if(sep[j] == N - 1):
            answer = True
            break
    if(answer == True):
        break

if(answer):
    print('Yes')
else:
    print('No')
```
